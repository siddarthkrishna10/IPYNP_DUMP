```
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

## **Creation of dictionaries and learning corpora**

## Exercise 1: 

### Question 1:

```
x = pd.read_csv('train.tags.txt', header=None, names=['tags'])
y = pd.read_csv('train.words.txt', delimiter="\n", header=None, names=['words'])

tags_words = pd.concat([x,y], ignore_index=False, axis=1)
tags_words['dictionary'] = tags_words['tags'].str.cat(tags_words['words'], sep ="|||") 

print(tags_words)
```

## Exercise 2:

### Question 1

```
v = CountVectorizer()

print(v.fit_transform(tags_words['tags']).todense())
print(v.get_feature_names())
```
